```python
from dotenv import load_dotenv
from agents import Agent, Runner, trace

load_dotenv()
```

```python
agent = Agent(name="Jokester", instructions="you are a joke teller", model="gpt-5-nano")
# you can use it with other vendor models as well

with trace("telling a joke"): #for tracing under that heading
    result = await Runner.run(agent, "tell me a joke about autonomous AI agents") #coroutine
    print(result)
```
